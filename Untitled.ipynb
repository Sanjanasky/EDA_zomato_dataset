```python
df_country = pd.read_excel('Country-code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on = 'Country Code', how ='left')
```

```python
final_df.head()
```

```python
final_df.columns
```

```python
country_name = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
##pie chart - Top 3 countries that uses zomato
plt.pie(country_val[:3],labels = country_name[:3],autopct = '%1.2f%%')
```

Obervation : Zomato Maximum records or transaction are from India after that USA and then United Kingdom

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = { 0:'Rating Count'})
```

```python
ratings
```

obersation 
1. When Rating is between 4.5 to 4.9 -->execellent
2. When Rating is between 4.0 to 3.4 -->very good
3. When Rating is between 3.5 to 3.9 -->good
4. When Rating is between 3.0 to 3.4 -->average
5. When Rating is between 2.5 to 2.9 -->average
6. When Rating is between 2.0 to 2.4 -->Poor


```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = "Aggregate rating" , y = "Rating Count" , data= ratings)
```

```python
sns.barplot(x = "Aggregate rating" , y = "Rating Count",hue = 'Rating color' , data= ratings,palette = ['blue','red','orange','yellow','green','pink'])
```

OBERSATION : 
1. Not Rated countis very high
2. Maximum number of rating are between 2.5 to 3.4

```python
sns.countplot(x = "Rating color", data = ratings, hue = 'Rating color',palette = [ 'blue','red','orange', 'yellow','green','pink'])
```
